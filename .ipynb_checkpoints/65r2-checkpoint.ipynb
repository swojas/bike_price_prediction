```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('BIKE DETAILS.csv')
```

```python
dataset.drop(['name'], axis=1, inplace=True)
```

```python
dataset['current_year'] = 2021
```

```python
dataset['diff_year'] = dataset['current_year'] - dataset['year']
dataset.drop(['year', 'current_year'], axis=1, inplace=True)
dataset.replace({'1st owner':1, '2nd owner':2, '3rd owner':3, '4th owner':4}, inplace=True)
```

```python
dataset = pd.get_dummies(dataset, drop_first=True)
```

```python
X = dataset.iloc[:, 1:].values
y = dataset.iloc[:, 0].values
```

```python
nan_col = dataset[dataset['ex_showroom_price'].isnull()]
nan_col = pd.DataFrame(nan_col)
nan_col.drop(['ex_showroom_price'], axis = 1, inplace = True)
nan_col = nan_col.values
non_nan = dataset[dataset['ex_showroom_price'].notnull()]
```

```python
non_nan_X = non_nan.drop(['ex_showroom_price'], axis=1).values
non_nan_X = pd.DataFrame(non_nan_X).values
non_nan_y = non_nan.iloc[:, -2].values
non_nan_y = pd.DataFrame(non_nan_y).values
```

```python
nan_col
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor(n_estimators = 100, random_state = 0)
rfr.fit(non_nan_X, non_nan_y)
```

```python
nan_pred = rfr.predict(nan_col) * 10000
```

```python
nan_pred = pd.DataFrame(nan_pred)
type(nan_pred)
print(len(nan_pred))
print(nan_pred)
```

```python
nan_col = pd.DataFrame(nan_col)
type(nan_col)
print(len(nan_col))
print(nan_col)
```

```python
result = pd.concat([nan_col, nan_pred], axis=1, ignore_index=True)
```
